```python
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
penguins = sns.load_dataset("penguins")
penguins
```

```python
sns.relplot(x="bill_length_mm", y ="bill_depth_mm", hue='species',style="island
```

```python
sns.relplot(x="bill_length_mm", y ="bill_depth_mm", hue='species',col="island", data=penguins)
```

```python
sns.relplot(x="bill_length_mm", y ="bill_depth_mm", 
hue='species',style="island", size="body_mass_g", sizes = (10,200),
data=penguins)
```

```python
flights = sns.load_dataset("flights")
flights
```

```python
sns.relplot(

    x="year",y="passengers", kind="line", data=flights, markers =True, dashes=False
)
```

```python
sns.relplot(

    x="year",y="passengers", kind="line", hue='month',data=flights, ci=None
)
```

```python
palette = sns.cubehelix_palette( n_colors=12, light=0.8)
sns.relplot(

    x="year",y="passengers", kind="line", hue='month',data=flights, ci=None,
    palette = palette, style ='month'
)
```

```python
sns.relplot(

    x="year",y="passengers", kind="line", col='month',data=flights, ci=None,
    col_wrap=4, linewidth=4, marker =True
)
```

```python
palette = sns.cubehelix_palette( n_colors=12, light=0.8)
g= sns.relplot(

    x="year",y="passengers", kind="line", hue='month',data=flights, ci=None,
    palette = palette, style ='month'
)
g.fig.autofmt_xdate()
```

```python
sns.catplot(x='species', y='body_mass_g', data=penguins)
```

```python
sns.catplot(x='species', y='body_mass_g', data=penguins, jitter=False)
```

```python
sns.catplot(x='species', y='body_mass_g', data=penguins, kind='swarm',
order = ['Gentoo','Adelie', 'Chinstrap'])
```

```python
sns.catplot(x='species', y='body_mass_g', data=penguins, kind='swarm',
col ="island", aspect =.7,
order = ['Gentoo','Adelie', 'Chinstrap'])
```

```python
sns.catplot(x='species', y='body_mass_g', data=penguins, kind='box',
col ="island", aspect =.7,
order = ['Gentoo','Adelie', 'Chinstrap'])
```
