```python
import seaborn as sns
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set(style="whitegrid")
```

```python
diamonds = sns.load_dataset("diamonds")
diamonds
```

```python
sns.catplot(x="cut", y ="price", kind = "boxen", data = diamonds)
```

```python
sns.catplot(x="cut", y ="price", kind = "violin", 
bw = .15, cut = 0, split = True, 
data = diamonds)
```

```python
sns.catplot(x="cut", y ="price", kind = "bar", data = diamonds)
```

```python
sns.catplot(x="cut", y ="price", kind = "bar", data = diamonds, hue ="color",
palette = "ch:.20")
```
